6. What are the differences between NumPy arrays and Python lists?
   - NumPy is the fundamental package for scientific computing in Python. Numpy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences. Numpy is not another programming language but a Python extension module. It provides fast and efficient operations on arrays of homogeneous data.
   - A Python list is a collection that is ordered and changeable. In Python, lists are written with square brackets.
   - The differences between NumPy arrays and Python lists are:
   - Python Lists
  - Element Overhead: Lists in Python store additional information about each element, such as its type and reference count. This overhead can be significant when dealing with a large number of elements.
  - Datatype: Lists can hold different data types, but this can decrease memory efficiency and slow numerical operations.
  - Memory Fragmentation: Lists may not store elements in contiguous memory locations, causing memory fragmentation and inefficiency.
  - Performance: Lists are not optimized for numerical computations and may have slower mathematical operations due to Python’s interpretation overhead. They are generally used as general-purpose data structures.
  - Functionality: Lists can store any data type, but lack specialized NumPy functions for numerical operations.
  - Numpy Arrays
  - Homogeneous Data: NumPy arrays store elements of the same data type, making them more compact and memory-efficient than lists.
  - Fixed Data Type: NumPy arrays have a fixed data type, reducing memory overhead by eliminating the need to store type information for each element.
  - Contiguous Memory: NumPy arrays store elements in adjacent memory locations, reducing fragmentation and allowing for efficient access.
  - Array Metadata: NumPy arrays have extra metadata like shape, strides, and data type. However, this overhead is usually smaller than the per-element overhead in lists.
  - Performance: NumPy arrays are optimized for numerical computations, with efficient element-wise operations and mathematical functions. These operations are implemented in C, resulting in faster performance than equivalent operations on lists.

15. How does NumPy handle multidimensional arrays?
    - NumPy handles multidimensional arrays (also called ndarrays) efficiently, making it a powerful tool for numerical computations in Python. For numpy, all data in any multi-dimensional array is actually stored in memory as a long 1D array (we will get back to this in the master lecture). The number of dimensions and shape of an array is actually only used to structure the data access in a certain way.
    - Here's how it works:
    -  Homogeneous Data: NumPy arrays store elements of the same data type (e.g., integers, floats, etc.) in a contiguous block of memory. This allows for faster access and operations compared to Python lists, which can store elements of different types.
    - Shape and Dimensions: NumPy arrays have a shape attribute that defines their dimensions. For example, a 2D array has a shape of (rows, columns). You can access and modify the shape using array.shape.
    - Indexing and Slicing: NumPy provides efficient ways to access and modify elements of multidimensional arrays using indexing and slicing. You can use integer indexing, boolean indexing, and slicing to extract specific elements or subarrays.
    - Broadcasting: NumPy supports broadcasting, which allows for operations between arrays with different but compatible shapes. This simplifies code and makes it easier to perform element-wise operations on arrays of different sizes.
    - Vectorized Operations: NumPy operations are vectorized, meaning they can be applied to entire arrays without the need for explicit loops. This results in significant performance improvements, especially for large datasets.
    - Memory Efficiency: NumPy arrays are more memory-efficient than Python lists, particularly for large datasets. This is because NumPy stores the data in a contiguous block of memory, avoiding the overhead associated with storing individual objects in Python lists.
    - C-API Integration: NumPy provides a C API that allows for efficient integration with low-level libraries written in C or Fortran. This enables NumPy to leverage optimized numerical routines and perform computations at speeds comparable to native C or Fortran code.

17. Explain the difference between apply() and map() in Pandas?
    - In Pandas, both apply() and map() are used to apply functions to data, but they differ in their scope and usage.
    - apply():
    - Scope: Can be used on both DataFrames and Series.
    - Functionality:
    - DataFrame: Applies a function along either the rows (axis=0) or columns (axis=1). The function can take the entire row/column as input.
    - Series: Applies a function to each element in the Series.
    - Flexibility: Can handle complex functions that require the entire row/column as input.
    - Example:
    - import pandas as pd
    - df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
    - Apply a function to each column
    - df.apply(lambda x: x.max() - x.min(), axis=0)
    - Apply a function to each row
    - df.apply(lambda x: x['A'] + x['B'], axis=1)
    - map():
    - Scope: Only applicable to Series.
    - Functionality: Applies a function to each element in the Series, replacing the original values with the output of the function.
    - Input: Accepts a function, a dictionary, or another Series to define the mapping.
    - Example:
    - s = pd.Series(['a', 'b', 'c'])
    - Map values using a dictionary
    - s.map({'a': 1, 'b': 2, 'c': 3})
    - Map values using a function
    - s.map(lambda x: x.upper())
    - Key Differences:
    - Data Structure:
    - apply() works on both DataFrames and Series, while map() only works on Series.
    - Function Input:
    - apply() can take functions that operate on entire rows or columns, while map() only applies functions element-wise.
    - Output:
    - apply() returns a Series or DataFrame, depending on the input and function. map() always returns a Series.
    - Which one to use?
    - Use apply() when you need to apply a complex function that requires the entire row or column as input.
    - Use map() for simple element-wise transformations on a Series.

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'A': [1, 2, 3, 4, 5],'B': [5, 4, 3, 2, 1], 'C': [2, 3, 4, 5, 6],'D': [5, 6, 7, 8, 9]}
df = pd.DataFrame(data)

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
